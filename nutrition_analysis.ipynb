### Handle Missing Values

```python
# Check for missing values
print(nutr_df.isnull().sum())
```

```python
# Fill missing values with 0 as it indicates the absence of that particular nutrient
nutr_df = nutr_df.fillna(0)
print(nutr_df.isnull().sum())
```

### Standardizing Column Names

```python
nutr_df.columns = nutr_df.columns.str.strip().str.replace('[()µ]', '', regex=True).str.replace(' ', '_')
nutr_df.columns = nutr_df.columns.str.replace('¾g', '3qtrg', regex=True)
nutr_df.columns = nutr_df.columns.str.replace('+', '_', regex=False)
print(nutr_df.columns)
```

### Check for Duplicates

```python
print(nutr_df.duplicated().sum())
```

```python
nutr_df.to_csv("data/nutrition_data.csv", encoding='utf-8', index = False)
```
